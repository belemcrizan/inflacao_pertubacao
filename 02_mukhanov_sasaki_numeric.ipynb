```python
import numpy as np
import matplotlib.pyplot as plt
from src.perturbation_equations import solve_mukhanov_sasaki
```

```python
eta_vals = np.linspace(-100, -0.1, 1000)  # tempo conforme (negativo)
k = 0.1  # modo k

eta, u_k = solve_mukhanov_sasaki(k, eta_vals)
```

```python
plt.plot(eta, u_k.real, label='Re(u_k)')
plt.plot(eta, u_k.imag, label='Im(u_k)')
plt.xlabel(r'$\eta$ (tempo conforme)')
plt.title(f'Evolução do modo Mukhanov–Sasaki para $k = {k}$')
plt.legend()
plt.grid(True)
plt.show()
```
